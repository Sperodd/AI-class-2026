```python
# 202435119 한여원
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# 202435119 한여원
# regression: linear regression, decision tree regression, random forest regression
df = pd.read_csv("D:\hanyeowon\Coding\pythonworks\deep\data\wine.csv")
print(df)
df = df.drop("Wine", axis=1)
print(df)

y = df["Alcohol"]
x = df.drop("Alcohol", axis=1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# 202435119 한여원
# linear regression
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()

linear_reg.fit(X_train, y_train)
y_pred = linear_reg.predict(X_test)

mse = metrics.mean_squared_error(y_test,y_pred)
print(f"mse = {mse}")

plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
# 202435119 한여원
# 색으로 비교
x = np.arange(len(y_test)) * 2

width = 0.6  # 바 두께

plt.figure(figsize=(25,6))

plt.bar(x - width/2, y_test, width, label='Actual', color='#0063B2')
plt.bar(x + width/2, y_pred, width, label='Predicted', color='#9CC3D5')

plt.xlabel("Sample index")
plt.ylabel("Value")
plt.title("Actual vs Predicted (Bar)")
plt.legend()

plt.show()
```

```python
# 202435119 한여원
# decision tree regression
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()

dt_reg.fit(X_train, y_train)
y_pred = dt_reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, y_pred)
print(f"mse = {mse}")

plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
```

```python
# 202435119 한여원
#random forest regression
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()

rf_reg.fit(X_train, y_train)
y_pred = rf_reg.predict(X_test)

mse = metrics.mean_squared_error(y_test, y_pred)
print(f"mse = {mse}")

plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```
