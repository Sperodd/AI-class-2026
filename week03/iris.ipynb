```python
# 202435119 한여원
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# 202435119 한여원
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X)
print(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# 202435119 한여원
from sklearn.linear_model import LogisticRegression

logit_reg = LogisticRegression(random_state=0, max_iter=1000)

logit_reg.fit(X_train, y_train)
y_pred = logit_reg.predict(X_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"accuracy={accuracy}")
print(f"{metrics.classification_report(y_test, y_pred)}")
disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
# 202435119 한여원
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=0)

dt_clf.fit(X_train, y_train)
y_pred = dt_clf.predict(X_test)

accuracy = metrics.accuracy_score(y_test, y_pred)

print(f"accuracy={accuracy}")
print(f"{metrics.classification_report(y_test, y_pred)}")
disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
# 202435119 한여원
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(random_state=0)

rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

accuracy = metrics.accuracy_score(y_test, y_pred)

print(f"accuracy={accuracy}")
print(f"{metrics.classification_report(y_test, y_pred)}")
disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
# 202435119 한여원
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors= 6)
knn_clf.fit(X_train, y_train)

y_pred = knn_clf.predict(X_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"accuracy for 6= {accuracy}")
```

```python
# 202435119 한여원
scores = []
k = int((len(X_train))+1)
for i in range(1, k):
    knn_clf = KNeighborsClassifier(n_neighbors= i)
    knn_clf.fit(X_train, y_train)

    y_pred = knn_clf.predict(X_test)

    accuracy = metrics.accuracy_score(y_test, y_pred)
    scores.append(accuracy)
plt.plot(range(1, k), scores)
plt.title("Accuracy for k")
plt.show()
```
